```python
# Libraries
import pandas as pd
from matplotlib import pyplot as plt
from sklearn import metrics
import numpy as np
```

<h4>1) Pressure value: 0.1 MPa</h4>

```python
# File input
filename1 = "input/pressure1.csv"
results_p1 = pd.read_csv(filename1, sep=";")
results_p1.tail()
```

```python
# Density values average, skipping 100000 first steps

# make Density column numeric
results_p1["Density"] = pd.to_numeric(results_p1["Density"])
# more info
# convert all columns of DataFrame: df = df.apply(pd.to_numeric)
# convert just columns "a" and "b": df[["a", "b"]] = df[["a", "b"]].apply(pd.to_numeric)

# skip 100000 first steps
skip = range(400, 1001) # generate a list 
av_density1 = results_p1["Density"].iloc[skip].mean()
print("Average value for density: {:.10f} kg/L".format(av_density1))
```

<h2>STEP 4</h2>

Error computation

```python
# Experimental data

# Source (Pádua): https://drive.google.com/drive/u/2/folders/1o_ANL2X8w3s3q2UkKrPkWo22dNZtmGul
# Temperature: 298.15 K
pressure_raw_1 = [0.1, 30.1, 50.05, 70, 100.1] # MPa
density_raw_1 = [687.82, 713.54, 726.83, 738.20, 753.07] # kg/m3
density_experimental_1 = [i * 0.001 for i in density_raw_1] # kg/L

# Source (Dymond): https://drive.google.com/drive/u/2/folders/1o_ANL2X8w3s3q2UkKrPkWo22dNZtmGul
# Temperature: 24.99 C
pressure_raw_2 = [0.1, 26.4, 50.7, 100.5, 200.4] # MPa
density_raw_2 = [687.82, 711.0, 728.2, 755.3, 792.2] # kg/m3
density_experimental_2 = [i * 0.001 for i in density_raw_2] # kg/L
# https://stackoverflow.com/questions/35166633/how-do-i-multiply-each-element-in-a-list-by-a-number/35166717

# Simulation data (See step 3)
# Temperature: 298 K
```

```python
# MAE - mean absolute error
def mae(pred_list, true_list):
    if len(pred_list) != len(true_list):
        raise Exception("Error: number of elements not match!")
    true_list, pred_list = np.array(true_list), np.array(pred_list)
    return np.mean(np.abs(true_list - pred_list))

# MAPE - mean absolute percentage error
def mape(pred_list, true_list):
    if len(pred_list) != len(true_list):
        raise Exception("Error: number of elements not match!")
    true_list, pred_list = np.array(true_list), np.array(pred_list)
    return np.mean(np.abs((true_list - pred_list) / true_list)) * 100
```

```python
# Error computation

print("Model's Error: {:.6f} kg/L = {:.6f} %"
      .format(mae(density_values, density_experimental_1), 
              mape(density_values, density_experimental_1)))
print("Model's Error: {:.6f} kg/L"
      .format(metrics.mean_absolute_error(density_experimental_1,
                                          density_values))) # using sklearn
```

```python
# Visualisation
# create a line chart, pressure on x-axis, density on y-axis 
plt.plot(pressure_values, density_values, 'g-', label='Simulation data')
plt.plot(pressure_raw_1, density_experimental_1, 'r-.', label='Experimetal data 1')
plt.plot(pressure_raw_2, density_experimental_2, 'b:', label='Experimetal data 2')

# add a title 
plt.title("Pressure vs Density") 

# add a legend
plt.legend(loc=9) # top center

# label each point
for label, pressure_v, density_v in zip(labels, pressure_values, density_values):
    plt.annotate(label, xy=(pressure_v, density_v), # put the label with its point
    xytext=(5, -5),                                 # but slightly offset
    textcoords='offset points') 

# add a label to the x-axis and y-axis 
plt.xlabel("Pressure MPa")
plt.ylabel("Density kg/L")
plt.grid(color='gray', linestyle='-', linewidth=0.1)
plt.savefig('density_comparison.png', dpi=300)
```
